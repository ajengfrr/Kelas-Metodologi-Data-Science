Bagaimana dengan data kategorikal?
Kita akan gunakan imputasi modus.

```python
from sklearn.impute import SimpleImputer
mobil = {'mobil':['Ford','Ford','Toyota','Honda',np.nan,'Toyota','Honda','Toyota',np.nan,np.nan,'Honda']}

data = pd.DataFrame(mobil)
```

```python
data
```

```python
imputasi = SimpleImputer(strategy='most_frequent')
```

```python
imputasi.fit_transform(data)
```

Bagaimana kalau kita butuh untuk mengatasi missing value untuk kategorikal dan numerikal secara bersamaan?
Kita bisa gunakan random sample

```python
#import Random Sample
from feature_engine.imputation import RandomSampleImputer

#buat ada yang ada missing value
data = {'Jenis Kelamin':['Laki-laki','Wanita','Laki-laki',np.nan,'Wanita','Laki-laki',np.nan,'Laki-laki',np.nan,'Wanita'],
        'Umur':[20, np.nan, 29, 32, np.nan, 52, 23, np.nan, 19, 26]}

df = pd.DataFrame(data)
df
```

```python
#buat imputer
imputer = RandomSampleImputer(random_state=25)

#fit kan
imputer.fit(df)

#ubah data
dftest = imputer.transform(df)
```
