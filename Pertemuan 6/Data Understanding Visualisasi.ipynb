```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

negara = ('Argentina', 'Brazil', 'Cina', 'Denmark', 'Estonia', 'Finlandia', 'Ghana', 'Hongaria', 'Indonesia', 'Jerman')
populasi = (45380000, 212600000, 1414830000, 5831000, 1200000, 5531000, 28100000, 975000, 273500000, 83240000)

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by = 'Population', inplace = True)

kor_x = np.arange(len(negara))
plt.bar(kor_x, df['Population'], tick_label = df['Country'])
plt.xticks(rotation = 90)
plt.ylabel('Populasi (juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show
```

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

negara = ('Argentina', 'Brazil', 'Cina', 'Denmark', 'Estonia', 'Finlandia', 'Ghana', 'Hongaria', 'Indonesia', 'Jerman')
populasi = (45380000, 212600000, 1414830000, 5831000, 1200000, 5531000, 28100000, 975000, 273500000, 83240000)

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by = 'Population', inplace = True)

kor_x = np.arange(len(negara))
warna = ['#0000FF' for _ in range(len(df))]
warna [-3] = '#00FF00'

plt.bar(kor_x, df['Population'], tick_label = df['Country'], color = warna)
plt.xticks(rotation = 90)
plt.ylabel('Populasi (juta)')
plt.title('Urutan Negara Ke-3 Terbesar Sesuai Data')
plt.show
```
